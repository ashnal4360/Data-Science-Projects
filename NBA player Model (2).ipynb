```python
import pandas as pd 
from sklearn.model_selection import train_test_split 
from sklearn.ensemble import RandomForestClassifier 
from sklearn.preprocessing import LabelEncoder 
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report 
```

```python
dataset = pd.read_excel(r"C:\Users\ashna\OneDrive\Documents\School\UMGC\Datasets\nba_logreg v2.xlsx")
dataset
#I originally was going to do a project on mental illness causing suicide but I decided on this topic because I found the 
#dataset more accessible. In hind sight I realized I should have used kaggle because there would have been a bigger repository
#that I could have worked with. 
```

```python
# dataset
# print("Sum of NULL values in each column. ") 
# dataset.fillna(0)
print(dataset.isnull().sum()) 
#I got rid of null values
```

```python
# dataset = dataset.replace('(NA)',0)
X = dataset.drop(['TARGET_5Yrs', 'Name'], axis = 1).values
#I dropped name from the dataset because it wasn't numerical. 
y = dataset[['TARGET_5Yrs']].values
```

```python
labelencoder_y = LabelEncoder() 
y = labelencoder_y.fit_transform(y) 
```

```python
X_train, X_test, y_train, y_test = train_test_split( 
              X, y, test_size = 0.3, random_state = 0)
```

```python
classifier=RandomForestClassifier() 
classifier=classifier.fit(X_train,y_train) 
predicted=classifier.predict(X_test) 
```

```python
print ('Confusion Matrix :') 
print(confusion_matrix(y_test, predicted)) 
print ('Accuracy Score :',accuracy_score(y_test, predicted)) 
print ('Report : ') 
print (classification_report(y_test, predicted)) 
```
